## Using API

```python
def get_musicbrainz_artist_data(artist_name):
    base_url = "https://musicbrainz.org/ws/2/"
    endpoint = "artist/"
    query_params = {
        "query": artist_name,
        "fmt": "json"
    }
    
    # Construct the full URL
    request_url = f"{base_url}{endpoint}"
    
    # Headers to mimic a browser request
    headers = {
        'User-Agent': 'AppName/1.0 ( YourContactInfo )',
        'Accept': 'application/json'
    }
    
    # Make the request
    response = requests.get(request_url, params=query_params, headers=headers)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        
        # Returning the full data for demonstration purposes
        # You might want to process this data further to fit your needs
        return data
    else:
        print("Failed to retrieve data")
        return None

# Example usage
artist_name = "Michael Jackson"
data = get_musicbrainz_artist_data(artist_name)

# Print the result
print(json.dumps(data, indent=4))
```
